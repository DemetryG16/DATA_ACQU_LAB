# Check with the professor all the parameters down here

```python
L_sun = 3.828 * 10**33                                # erg/s
M_sun = 1.988 * 10**33                                # gram
gal_dist = 5.545 *(10**6)* 3.0857*10**(18)     # 5.545 Â± 0.260    # We can have it directly from NED/NASA (It was in Mpc and now it's in cm)
gal_rad = 26.77*(10**3)* 3.0857*(10**(18))/2   # The galaxy radius from NED/NASA in cm               
area = 4*np.pi*(gal_dist**2)   # Area of the Galaxy sphere at gal_dist

flux_r = 1.299878e-08              # total flux from surface brightness file in cgs, the attenuation is removed.
flux_g = 3.3538572118465e-08       # total flux from surface brightness file in cgs, the attenuation is removed.
f_nu_r = flux_r/1.1e14
f_nu_g = flux_g/1.8e14
flux_r_a = 2.925e-10               # total flux from surface brightness file in cgs, the attenuation is not removed.
flux_g_a = 2.119e-10               # total flux from surface brightness file in cgs, the attenuation is not removed.

a_i = -0.222
b_i = 0.864
a_r = -0.306
b_r = 1.097
a_g = -0.499
b_g = 1.519 

color = -2.5*np.log10(flux_g/flux_r)

A_r = 4.12
A_g = 5.5

gal_mass_r = (flux_r*(10**(A_r/2.5))*area/L_sun)*10**(a_r + b_r*color)    # in units of solar mass 
gal_mass_g = (flux_g*(10**(A_g/2.5))*area/L_sun)*10**(a_g + b_g*color)
print(f'{gal_mass_r:.2e}')
print(f'{gal_mass_g:.2e}')
color
```

# Mass Error ($\sigma_{M}$)

To derive the error, use the following link: https://en.wikipedia.org/wiki/Propagation_of_uncertainty.
* Remember that you need only to use the table of the functions in the link and go step by step for each function.

```python
E_flux_g = 1.171e-11
E_flux_r = 3.477e-12


E_M_r = gal_mass_r*np.sqrt((E_flux_r/flux_r)**2+((E_flux_r/flux_r)**2+(E_flux_g/flux_g)**2)*(b_r*2.5)**2)
E_M_g = gal_mass_g*np.sqrt((E_flux_g/flux_g)**2+((E_flux_r/flux_r)**2+(E_flux_g/flux_g)**2)*(b_r*2.5)**2)
```

```python
print(f"{E_M_r:.3e}")
```

```python
print(f"{E_M_g:.3e}")
```
